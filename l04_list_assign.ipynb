```python
# Question 7: Find the second largest number in [45, 67, 23, 89, 12, 34, 78]
numbers = [45, 67, 23, 89, 12, 34, 78]

largest = numbers[0]
second_largest = numbers[0]

for num in numbers:
    if num > largest:
        second_largest = largest
        largest = num
    elif num > second_largest and num != largest:
        second_largest = num

print(second_largest)
```

```python
# Question 10: Create a list of prime numbers between 1 and 50
primes = []

for num in range(2, 51):
    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        primes.append(num)

print(primes)
```

```python
# Question 15: Transpose the matrix [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
matrix = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]

transpose = []

for i in range(len(matrix[0])):
    row = []
    for j in range(len(matrix)):
        row.append(matrix[j][i])
    transpose.append(row)

print(transpose)
```

```python
# Question 16: Find the maximum value in each sublist of [[1, 5, 3], [9, 2, 7], [4, 8, 6]]
lists = [[1, 5, 3], [9, 2, 7], [4, 8, 6]]
max_values = []

for sublist in lists:
    max_val = sublist[0]  # assume first element is max
    for num in sublist:
        if num > max_val:
            max_val = num
    max_values.append(max_val)

print(max_values)
```

```python
# Question 18: Find the sum of all elements in 3D list [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
list_3d = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]

total_sum = 0

for matrix in list_3d:         # 2D lists
    for sublist in matrix:     # 1D lists
        for num in sublist:    # elements
            total_sum += num

print(total_sum)
```

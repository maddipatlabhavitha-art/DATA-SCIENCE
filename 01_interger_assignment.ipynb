```python
# Question 5: Calculate the sum of digits in number 12345
number=12345
sum=0
while number>0:
  digit=number%10
  sum += digit
  number //= 10
print(sum)

```

```python
# Question 6: Check if 97 is a prime number

num = 97
is_prime = True

if num <= 1:
    is_prime = False
else:
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

```python
# Question 7: Find the factorial of 8
num=8
fact=1
for i in range(1,num+1):
  fact=fact*i
print(fact)
```

```python
numbers = [15, 23, 31, 42, 56]

sum_nums = 0
for num in numbers:
    sum_nums += num

average = sum_nums / len(numbers)
print("Average:", average)
```

```python
# Question 9: Find the greatest common divisor (GCD) of 48 and 36

a = 48
b = 36

gcd = 1
for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        gcd = i

print("GCD of", a, "and", b, "is:", gcd)
```

```python
# Question 10: Calculate the sum of first 20 odd numbers
sum_odd = 0
count = 0
num = 1

while count < 20:
    sum_odd += num
    num += 2      # move to next odd number
    count += 1

print("Sum of first 20 odd numbers:", sum_odd)
```
